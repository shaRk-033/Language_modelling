```python
def train_epochs(y, weights, epochs = 100, lr = 10):
    for i in range(epochs):
        
        #forward prop
        x_emb = F.one_hot(xt, num_classes = 27).float()
        probs = (x_emb @ weights).exp()
        probs = probs/probs.sum(1, keepdims = True)
        loss = -probs[torch.arange(num), yt].log().mean() + 1*(weights**2).mean()
        print(loss)
        #backward prop
        loss.backward()
        
        with torch.no_grad():
            weights_copy = weights.clone() 
            weights_copy -= lr * weights.grad
            weights.copy_(weights_copy)
            weights.grad.zero_()         
        
```
